** Fill missing values ** 

```python
df['experience'].fillna('zero',inplace=True)
df
```

**Convert word to number** :
              
               It is process of converting textual numbers(words) into numerical form.

```python
from sklearn import linear_model
from word2number import w2n
```

```python
df.experience=df.experience.apply(w2n.word_to_num)
```

```python
df
```

** Fill missing values using aggregat function **

```python
import math
mean=math.floor(df['test_score(out of 10)'].mean())
mean
```

```python
df['test_score(out of 10)'].fillna(mean,inplace=True)
```

```python
df
```

** Segregate date into Indepndent variable(x) ane Dependent variable(y) **

```python
df.columns
```

```python
predict=['experience', 'test_score(out of 10)', 'interview_score(out of 10)']
x=df[predict]
y=df['salary($)']
```

** Spliting dataset into training and testing ** 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

 ** Fitting the simple linear regression Model to the training dataset and Prediction of trainig set Result. **

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict([[4,5,6]])
```

** Training Result **

```python
reg.score(x_train,y_train)
```

** Testing Result **

```python
reg.score(x_test,y_test)
```
