## Multiple Linear Regression (MLR)

Multiple linear regression (MLR) is a statistical method that predicts a dependent variable using multiple independent variable.It's an extension of simple linear regression ,which uses only one independent variables.

dummies -It is refer to dummy variables(also known us one - hot encoding).They are used to categorical data into a 

numerical format that can be used in machine learning models.

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
merge=pd.concat([df,dummies],axis='columns')
merge
```

```python
final=merge.drop(['Car Model'],axis='columns')
final
```

```python
final.columns
```

```python
** Segregate dataset into Independent variable(x) and Dependent variable (y)
```

```python
predictors=['Mileage','Age','Audi','BMW X5','Mercedez Benz ','Toyota']
x=final[predictors]
y=final['Sell Price']
x.shape,y.shape
```

** Spliting dataset into training and testing **

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=43)
```

 ** Fitting the simple linear regression Model to the training dataset and Prediction of trainig set Result.**

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg
```

```python
model=reg.fit(x_train,y_train)
```

** Testing Result **

```python
model.score(x_test,y_test)
```

** Training Result **

```python
model.score(x_train,y_train)
```
