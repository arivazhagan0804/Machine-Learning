** Segregate data into Independent variable(x) and Dependent variable **

```python
df.head(1)
```

```python
df.columns
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,5:].values
```

```python
x
```

```python
y
```

**Spliting data into Trainig ans Testing**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

**Fitting Linear Regression Model to training dataset and prediction of traing set Result**

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

**Fitting the Model**

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
