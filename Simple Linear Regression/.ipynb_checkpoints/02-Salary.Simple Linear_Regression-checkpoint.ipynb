## Linear Regression:

**Linear regression is also type of supervised machine learning  algorithm that learn from the labelled datasets and maps the data points with most optimized linear 
function which can be used for prediction on new datasets.It computes the linear relationship between the dependent variable and one more independent feature by fitting a linear equation with observed data.It predicts the continuous output variables based on the independent input variable.**

** import the libraries **

```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
```

** Reating the Dataset **

```python
df=pd.read_csv(r'C:\Users\ELCOT\Desktop\Salary_Data.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

** Segregrate the dataset into Independent Varible (x) and Dependent Variable (Y) **

iloc-It workby looking for the position in the dataframe.

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1].values
x
```

```python
print(y)
```

Splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

Fitting the simple linear regression Model to the training dataset and Prediction of trainig set Result.

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
x_pred=reg.predict(x_train)
```

```python
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train,x_pred)
plt.title('Salary vs Experience(Trainig Dataset)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary(inRupees)')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='green')
plt.plot(x_train,x_pred)
plt.title('Salary vs Experience(Trainig Dataset)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary(inRupees)')
plt.show()
```
