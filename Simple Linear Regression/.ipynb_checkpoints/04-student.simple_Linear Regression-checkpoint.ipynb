## Student Dataset

Data information:

* Hours: Time spent on one activity(e.g.,lecture/study)

* sHours: Possibly secondary hours(e.g.,self-study or supplementary)

* Marks : Student Score

I will be working with Student Dataset these excercise.For now,I will be focus on using supervised machine learning regression algorithm for prediction tasks.

** import libraries **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

** Read dataset **

```python
df=pd.read_csv(r'C:\Users\ELCOT\Desktop\02Students - lab (1).csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
x=df.iloc[:,:-2].values
y=df.iloc[:,2:].values
x
```

```python
y
```

Spliting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

Fitting the simple linear regression Model to the training dataset and Prediction of trainig set Result

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
x_pred=reg.predict(x_train)
y_pred=reg.predict(x_test)
```

```python
x_pred.mean()
```

```python
y_pred.mean()
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,x_pred)
plt.title('Hour vs Student mark')
plt.xlabel('Hour')
plt.ylabel('Marks')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,x_pred)
plt.title('Hour vs Student mark')
plt.xlabel('Hour')
plt.ylabel('Marks')
plt.show()
```
