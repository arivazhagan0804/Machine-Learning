** Import Libraries **

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Reading the  Dataset**

```python
df=pd.read_csv(r'C:\Users\ELCOT\Desktop\csv file\Machine Learning\LInear Regression\Simple Linear Regression\ad_spend_vs_sales.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

** Segregate data into Independent variable(x) and Dependent variable(y). **

iloc-It's workby looking for the position in the dataframe.

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1:].values
x
```

```python
y
```

** Spliting dataset into traing and testing **

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

** Fitting the simple linear regression Model to the training dataset and Prediction of trainig set Result.**

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
x_pred=reg.predict(x_train)
x_pred
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,x_pred,color='red')
plt.title('Ad spend vs Sales(Training data)')
plt.xlabel('Ad Spend ($1000s)')
plt.ylabel('Sales ($1000s)')
plt.show()
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_train,x_pred,color='red')
plt.title('Ad spend vs Sales(Test dataset)')
plt.xlabel('Ad Spend($1000s)')
plt.ylabel('Sales($1000s)')
plt.show()
```
