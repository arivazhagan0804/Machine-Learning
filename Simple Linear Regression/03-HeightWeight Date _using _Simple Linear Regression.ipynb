** Segregate the Dataset into Independent variable(x) and Dependent variable(y) **

** iloc-It's workby looking for the position in the dataframe **

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1:].values
x
```

```python
y
```

```python
** Spliting data into Training and Testing **
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

** Fitting the linear regression Model to the training dataset and Prediction of trainig set Result. **

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

```python
model=reg.fit(x_train,y_train)
model
```

```python
x_pred=model.predict(x_train)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,x_pred)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs Height (Training Data)')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,x_pred)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs Height (Test Data)')
plt.show()
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
